# Visualização de Dados

### Titanic

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('D:\\Programação\\Estudos\\Udemy\\Data Science e ML\\05 - Visualizacao de Dados\\titanic\\train.csv')
data
```

```python
data.columns
```

```python
data.columns = ['IdPassageiro', 'Sobrevivente', 'Classe', 'Nome', 'Sexo', 'Idade', 
                'IrmaosConjuge','PaisFilhos', 'Bilhete', 'Tarifa', 'Cabine', 'Embarque']
data.columns
```

```python
data['Sexo'].replace({'male': 'homem', 'female': 'mulher'}, inplace= True)
data['Cabine'] = data['Cabine'].apply(lambda x: x[0] if pd.notna(x) else np.nan)
data
```

### Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
 
```

```python
plt.hist(data['Idade'].dropna(), bins=30)
plt.title('Histograma das Idades')
plt.ylabel('Pessoas')
plt.xlabel('Idade')
plt.show()
```

```python
data['Classe'].unique()
```

```python
plt.hist(data['Classe'])
plt.title('Distribuição das Classes')
plt.ylabel('Pessoas')
plt.xlabel('Classes')
plt.show()
```

### Subplots - 1ª Forma

```python
plt.figure(figsize=(6,3), dpi=150)
plt.subplot(1,2,1) #1 linha, 2 colunas, plot 1
plt.hist(data['Idade'].dropna())
plt.title('Histograma das Idades')

plt.subplot(1,2,2) # 1 linha, 2 colunas, plot 2
plt.hist(data['Classe'])
plt.title('Distribuição das Classes')

plt.show()
```

### Subplots - 2ª Forma

```python
f, ax = plt.subplots(1,2, figsize=(6,3), dpi=150) # 1 linha, 2 colunas
ax[0].hist(data['Idade'].dropna())
ax[0].set_title('Distribuição das Idades')

ax[1].hist(data['Classe'])
ax[1].set_title('Distribuição das Classes')

plt.tight_layout()
plt.show()
```

### Seaborn

### pie e countplot

```python
f, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].pie(data['Classe'].value_counts(), explode=[0.02,0.02,0.02], autopct='%0.2f%%')
ax[0].set_ylabel('')

sns.countplot(x = 'Sobrevivente', data=data, ax=ax[1])
ax[1].set_ylabel('')
```

### countplot  e hue

```python
sns.countplot(x = 'Sexo', hue='Sobrevivente', data=data)
```

### boxplot

```python
plt.figure(figsize=(5,3), dpi=100)
sns.boxplot(x='Classe', y='Idade', hue='Sexo', data=data)
```

### violinplot

```python
plt.figure(figsize=(5,3), dpi=100)
sns.violinplot(x='Classe', y='Idade', hue='Sexo', data=data)
```

### FacetGrid com hist

```python
g = sns.FacetGrid(data, col='Sobrevivente')
g.map(plt.hist, 'Idade', bins=18)
```

### FacetGrid com barplot

```python
g = sns.FacetGrid(data, row='Embarque', col='Sobrevivente')
g.map(sns.barplot, 'Sexo', 'Tarifa', alpha=0.8)

fig=plt.gcf()
fig.set_size_inches(10,6)
```

### FacetGrid com distplot

```python
g = sns.FacetGrid(data, col='Sobrevivente', row='Classe', height=1.8, aspect=2.5)
g.map(sns.distplot, 'Idade', bins=20)
```

### catplot do tipo "point"

```python
sns.catplot(x = 'Embarque', y = 'Sobrevivente', data=data, kind='point')
fig=plt.gcf()
fig.set_size_inches(6,3)
```

### FacetGrid com pointplot

```python
g = sns.FacetGrid(data, row='Embarque', height=1.8, aspect=2.2)
g.map(sns.pointplot, 'Classe', 'Sobrevivente', 'Sexo')
g.add_legend()
```

### heatmap

```python
data.isnull().sum()
```

```python
sns.heatmap(data.isnull(), yticklabels=False, cbar=False, cmap='magma')
```

### relplot

```python
sns.relplot(x='Tarifa', y='Idade', hue='Sexo', data=data, size='Tarifa', sizes=(40,400),
           alpha=0.8, palette='magma')
plt.show()
```

### scatterplot

```python
sns.scatterplot(x='Idade', y='Tarifa', hue='Embarque', size='Idade',
               palette='magma', data=data, sizes=(10, 200))
```

### pairplot

```python
sns.pairplot(data[['Tarifa', 'Idade', 'Classe', 'Sexo', 'Embarque','Sobrevivente']], hue='Classe')
```

### jointplot tipo "kde"

```python
sns.jointplot(x='Idade', y='Tarifa', data=data, kind='kde', color='g')
```

### jointplot tipo "reg"

```python
sns.jointplot(x='Idade', y='Tarifa', data=data, kind='reg', color='g')
```
